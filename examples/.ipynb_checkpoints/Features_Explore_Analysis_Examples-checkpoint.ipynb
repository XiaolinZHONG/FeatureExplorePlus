#### Feature Explore Tool 分析变量分布

```python
features_list = df.drop(['label'], 1).columns
features_list
```

```python
fet = fep.FeatureExplore()  # 默认不使用决策树分BIN
fet.feature_exp_plots(
    data=df, target_col='label', features_list=['col1'])  # 这里可以批量传入变量
```

对于上图的分析：
+ 该变量缺失率很高，大部分值都是缺失值。需要排查原始数据逻辑是否有问题。
+ 该变量和标签的走势关系是倒U型，是一个较为理想的变量。上面的三个BIN 可以拆分作为二个变量输入给线性模型。
+ 左图背后的直方图说明数据在每一个BIN 上的数量。用来判断数据分布是否正常，可以和右图的原始分布作比较。
+ 基于该变量的策略制定：如果变量的值0那么出对应的策略，因为其有10%以上的概率会“成功”。

```python
fet.tree_split=False
fet.feature_exp_plots(
    data=df.fillna(-900), data_test=df2.fillna(-900), target_col='label', features_list=['col2'],bins=10)
```

对于上图的分析：
1. 上图加入了测试数据，一般情况下的测试数据集在生产上都是跨时间的，既是几天之后的数据。测试数据的引入可以更有效的分析变量
    + 首先训练数据和测试数据整体上都是大约50%的缺失值，50%的有值，其中有值得区间都在[0-1]之间。说明数据没有问题
    + 训练数据和测试数据之间的走势基本上完全不一致，说明该变量随着时间的变化不是很稳定，也可能是数据抽样不科学导致的。
    + 改变来那个走势的不一致和走势混乱程度决定了，该变量不适合作为特征，因为不容易去平滑走势（平滑操作一般是合并BIN等，用户可以自己试下，如果调小BINS数走势变好，就是需要平滑），除非使用深度学习模型等高级模型。
    + 下面可以使用基于最优分BIN法的方式划分BIN 看下数据的分布和走势。
2. 上图无论是训练数据还是测试数据，在BIN上的标签AVG都有大于0.1的，一般在生产环境下很少出现这种变量，建议排查数据看下是否有数据穿越的风险。当然如果真的是标签泄露，AVG值会非常的高。

#### 使用决策树分BIN

```python
fet.tree_split = True
# 目前版本 还无法直接处理NAN值，需要自己处理,一般填 较大的负值，便于拉开和正常值的距离。
fet.feature_exp_plots(
    data=df.fillna(-900),
    data_test=df2.fillna(-900),
    bins=5,
    target_col='label',
    features_list=['col2'])
```

对于上图的分析:
+ 通过减少BIN 数来让决策树自动平滑区间。使得训练集和测试集的数据走势一致，避免了过拟合的现象。
+ 如果不通过决策树来自动平滑，使用默认的等频分BIN 来平滑，也可以达到相同的效果。
+ 决策树分BIN 并不一定比等频分BIN效果好，对于不同的数据集和不同的BIN 有很大的差异，需要用户自行斟酌。大部分情况下决策树分BIN效果更优。

#### 调用决策树分BIN 方法得到区间值

```python
fet.get_tree_bins(df.fillna(-900),features_list=['col2'],target_col='label',bins=5)
```

返回每一个变量对应的决策树分BIN的区间分布，可以直接应用于 TF 等

#### 调用变量变化趋势

```python
fet.tree_split=True
fet.feature_trend_stats(
    data=df.fillna(-900),
    data_test=df2.fillna(-900),
    features_list=['col2'],
    target_col='label',
    bins=5)
```

```python
fet.tree_split = False
fet.feature_exp_plots(
    data=df.fillna(-900),
    data_test=df2.fillna(-900),
    target_col='label',
    features_list=['col2'],
    bins=5)
fet.feature_trend_stats(
    data=df.fillna(-900),
    data_test=df2.fillna(-900),
    features_list=['col2'],
    target_col='label',
    bins=5)
```

1. 基于决策树的分BIN 可以使得训练集和测试集的走势区域完全一致，其对应的PSI值（群体稳定性指标）也非常的低。一般认为变量的 PSI 低于0.01 就认为其为稳定变量，根据不同的业务场景变化。
2. 基于等频分BIN 的变量的测试集和训练集在最后一个BIN 走势相反，影响了变量的稳定性，进而影响模型的线上效果。而其PSI 值也低于0.01，这是由于其大部分的走势(分布)是一致的。如果业务上能接受，也可以上线使用。
3. Trend_correlation 和 PSI 指标可以用于变量的监控也可以用于模型分的监控，一般情况下，线上的监控都是计算当前值和模型训练时的值得一致性，PS:也有观察当前值和昨天或者前天的一致性，这种观察意义不如上面的。
